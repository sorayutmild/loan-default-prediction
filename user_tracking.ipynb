```python
def end_loan_date(years, months, days):
    years = np.asarray(years) - 1970
    months = np.asarray(months) - 1
    days = np.asarray(days) - 1 
    types = ('<M8[Y]', '<m8[M]', '<m8[D]')
    vals = (years, months, days)
    return sum(np.asarray(v, dtype=t) for t, v in zip(types, vals)
               if v is not None)
```

```python
trans_after_loan = pd.read_csv('data/trans_after_loan.csv')
trans_after_loan = pd.merge(trans_after_loan, loan_df[['account_id', 'duration']], how='left', on=['account_id'], validate='many_to_one')
trans_after_loan['end_loan_date'] = end_loan_date(years=pd.DatetimeIndex(trans_after_loan['loan_date']).year, 
                                                months=pd.DatetimeIndex(trans_after_loan['loan_date']).month + trans_after_loan['duration'],
                                                days=pd.DatetimeIndex(trans_after_loan['loan_date']).day)
trans_after_loan.head()
```
